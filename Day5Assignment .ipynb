```
 list = [0,1,2,10,4,1,0,56,2,0,1,3,0,56,0,4]
list.sort()
a=list.count(0)
print("Sorted list:",list[a:]+list[:a])
```

```
list1=[1,5,3,2,4]
list2=[7,10,8,6,9]
list3=list1+list2
list4=[]
for num in range(min(list3),max(list3)+1):
    if num in list3:
        list4.append(num)
print("Sorted list:",list4)
```
